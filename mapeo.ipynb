El mapeo del plano de poincaré para el escilador armónico

$y\prime \prime + w_0 ^2 sin(y)=g(t) $ 

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])])
```

```python
for i in range(150):
    y0,u0 = pi*(2*rand()-1),0
    t,sol = rk4 (osc,[y0,u0],[0,10.])
    plot(sol[:,0],sol[:,1])
zoom = pi
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()
```
